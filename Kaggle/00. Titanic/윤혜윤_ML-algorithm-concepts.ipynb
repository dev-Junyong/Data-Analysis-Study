##  kNN

최근접이웃 알고리즘
1. Assign **K** a value (홀수가 일반적임)
2. Find K items with **lowest distances** to new point
> distance between two points -> **피타고라스** 정의
3. Assign the majority class as our prediction for the class of new point

> K를 3으로 설정 시 Action Movie로 판단

## Decision Tree

Decision Tree 알고리즘
1. Define Problem - Collect training data
2. Extract Data (분류) - Build a tree
3. Deploy machine
4. Test with test data

Choose best feature to split -> 엔트로피가 감소하는 방향으로 학습을 진행.
1. Entropy - 불순도. 확률 변수의 불확실성을 수치로 나타낸 것
> measure of messiness
2. Information Gain 
> decrease in entropy after a dataset is split on an attrebute = 분할 전 엔트로피 - 분할 후 엔트로피

 ## SVM

linearly separable
1. margin
2. support vector - margin과 가장 가까운 데이터
* SVM only uses support vector for prediction
---

not linearly separable
* mapping lower dimension to high dimension -> 계산량이 너무 많아짐 -> Kerner trick 방법쓰면 됨 


**SVM Parameter**
1. **Cost** - 마진과 decision boundary 간격을 어느정도로?
* when cost is small - train error 확률 높아짐 but 성능 높
* when cost is high - error은 적어짐 but 성능 낮아질 수 있음.


2. **Gamma** - train 데이터 하나 당 영향을 끼치는 범위를 조절하는 변수

Gamma가 높을 수록 범위가 확연히 줄어듦

-> find **optimal parameters** - Grid Search
